### A simple text embedding call, made through SAP BTP gen aicore

```python
# Load the library
from genai_wrapper.wrapper import GenAIWrapper, ChatObject
```

```python
# Load the config file and instantiate the Wrapper
gen_ai = GenAIWrapper( config_file="ai_core.json" )
```

```python
# Create the chat object
chat = ChatObject()
# Invoke the text embedding
embed = gen_ai.embedding("text-embedding-ada-002", "Hello World!")
# Print the embedding vector
print(embed)
```
